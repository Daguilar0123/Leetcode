```python
from typing import List

class Solution:
    def plusOne(self, digits: List[int]) -> List[int]:
        # calculate length of digits/how many diguts?

        # digits = [9]
        l = len(digits)
        # l = 1

        # initialize variable for the rightmost digit
        end = l - 1
        print(f"end = l - 1")

        # loop through digits from right to left
        # starts at 'end', then takes '-1' steps,
        # until it reaches '-1' exclusive, which is '0'. It ends
        # at index '0'
        for index in range(end, -1, -1):
            # iteration 'end':

            total = digits[end] + 1
            print(f"iteration: {index}")
            print(f"index: {index}; digits = {digits}")
            print(f"total = digits[end] + 1")
            print(f"total = digits[{end}] + 1")
            print(f"total = {digits[end]} + 1 = {digits[end]+1}\n")

            print(f"if total <= 9:\n")
            if total <= 9:
                print(f"true")
                digits[end] = total
                print(f"digits[end] = total\n")
                return digits

            digits[end] = 0

            if end == 0:
                return [1] + digits

            end -= 1

        return digits
```

```python


input_digits = [9,9,9]
expected_output = [1,0,0,0]
solution_instance = Solution()
# Create a copy of input_digits to avoid in-place modification affecting the printout
actual_output = solution_instance.plusOne(list(input_digits))

print(f"Input: {input_digits}")
print(f"Expected Output: {expected_output}")
print(f"Actual Output: {actual_output}")
```
