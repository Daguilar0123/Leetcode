```python
from typing import List

class Solution:
    def plusOne(self, digits: List[int]) -> List[int]:
        # calculate length of digits/how many diguts?

        # digits = [9]
        l = len(digits)
        # l = 1

        # initialize variable for the rightmost digit
        end = l - 1
        # end = 1-1=0

        # loop through digits from right to left
        # starts at 'end', then takes '-1' steps,
        # until it reaches '-1' exclusive, which is '0'. It ends
        # at index '0'

        # iteration 'end': for index in range(0,-1,-1):
        # this means we start at '0', then step '-1' (or step "backwards") until
        # '-1' exclusive.
        # However, since '-1' exclusive, coming from the right (i.e. '-1' step), is '0',
        # then the loop ends after the first iteration.
        # digits = [9]
        for index in range(end, -1, -1):
            # iteration 'end':
            # total = digits[0] + 1
            # total = 9 + 1 = 10
            total = digits[end] + 1

            if total <= 9:
                digits[end] = total
                return digits

            # digits[0] = 0
            digits[end] = 0

            if end == 0:
                return [1] + digits
            # end = 0 - 1 = -1
            end -= 1
            # digits = [0]
            # end = -1
        return digits
```

```python
solution_instance = Solution()

input_digits = [9,9,9]
expected_output = [1,0,0,0]

actual_output = solution_instance.plusOne(input_digits)

print(f"Input: {input_digits}")
print(f"Expected Output: {expected_output}")
print(f"Actual Output: {actual_output}")
```
